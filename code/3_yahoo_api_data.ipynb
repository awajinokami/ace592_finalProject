## To match the timestamp of bitcoin's sentiment score, we work on the time stamp and convert it to UTC in below

```python
#use 'dt' as datatimes
```

```python
# For nz_1m
doge_1m =pd.read_csv('./4_yahoo_api_data/nz_1m.csv')
nz_1m['dt']=pd.to_datetime(nz_1m.Datetime,format='%Y-%m-%d %H:%M')
nz_1m['dt'] = nz_1m['dt'] - timedelta(hours=12)
nz_1m['dt']=nz_1m['dt'].dt.tz_localize(None)
nz_1m['times'] = pd.to_datetime(nz_1m['dt'])
nz_1m['times'] = nz_1m['times'].dt.time
mask =(nz_1m['times'].astype(str) > '22:00')
nz_1m=nz_1m.loc[mask]
nz_1m.to_csv('./4_yahoo_api_data/nz_1m_new.csv')
```

```python
# For doge_1m
doge_1m =pd.read_csv('./4_yahoo_api_data/doge_1m.csv')

doge_1m['dt']=pd.to_datetime(doge_1m.Datetime,format='%Y-%m-%d %H:%M')
doge_1m['dt'] = doge_1m['dt'] - timedelta(hours=1)
doge_1m['dt']=doge_1m['dt'].dt.tz_localize(None)
doge_1m['times'] = pd.to_datetime(doge_1m['dt'])
doge_1m['times'] = doge_1m['times'].dt.time
mask =(doge_1m['times'].astype(str) > '22:00')
doge_1m=doge_1m.loc[mask]
doge_1m.to_csv('./4_yahoo_api_data/doge_1m_new.csv')
```

```python
#For eth_1m
eth_1m=pd.read_csv('./4_yahoo_api_data/1m_eth.csv')

eth_1m['dt']=pd.to_datetime(eth_1m.Datetime,format='%Y-%m-%d %H:%M')
eth_1m['dt'] = eth_1m['dt'] - timedelta(hours=1)
eth_1m['dt']=eth_1m['dt'].dt.tz_localize(None)
eth_1m['times'] = pd.to_datetime(eth_1m['dt'])
eth_1m['times'] = eth_1m['times'].dt.time
mask =(eth_1m['times'].astype(str) > '22:00')
eth_1m=eth_1m.loc[mask]
eth_1m.to_csv('./4_yahoo_api_data/eth_1m_new.csv')
```
